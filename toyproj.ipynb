```python
import os
import re
import unicodedata
import numpy as np
import pandas as pd
```

```python
def normalize_text(s):
    return unicodedata.normalize("NFC", str(s)).strip()

def find_file_by_keyword(file_list, keyword):
    keyword = normalize_text(keyword)
    for f in file_list:
        if keyword in normalize_text(f):
            return f
    raise FileNotFoundError(f"'{keyword}' 포함 파일을 찾지 못했습니다.")
```

```python
def clean_station_text(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    x = re.sub(r'\s+', '', x)
    return x

station_master_merge['station_name'] = station_master_merge['station_name'].apply(clean_station_text)
station_master_merge['line_name'] = station_master_merge['line_name'].apply(clean_station_text)

subway_time['station_name'] = subway_time['station_name'].apply(clean_station_text)
subway_time['line_name'] = subway_time['line_name'].apply(clean_station_text)
```

```python
def filter_valid_seoul_coords(df, lat_col='lat', lon_col='lon'):
    out = df.copy()
    out[lat_col] = pd.to_numeric(out[lat_col], errors='coerce')
    out[lon_col] = pd.to_numeric(out[lon_col], errors='coerce')

    before = len(out)

    out = out.dropna(subset=[lat_col, lon_col])
    out = out[
        (out[lat_col].between(37.40, 37.72)) &
        (out[lon_col].between(126.75, 127.20))
    ].copy()

    after = len(out)
    print(f"removed invalid coord rows: {before - after}")
    return out

shelter = filter_valid_seoul_coords(shelter)
store = filter_valid_seoul_coords(store)
hospital = filter_valid_seoul_coords(hospital)
station_master_merge = filter_valid_seoul_coords(station_master_merge)
subway_time_merged = filter_valid_seoul_coords(subway_time_merged)
```

```python
# 위경도 격자 키 만들기
from pyproj import Transformer

transformer = Transformer.from_crs("EPSG:4326", "EPSG:5179", always_xy=True)

def add_grid_key(df, lon_col='lon', lat_col='lat', grid_size=500):
    out = df.copy()

    x, y = transformer.transform(out[lon_col].values, out[lat_col].values)
    out['x_m'] = x
    out['y_m'] = y

    out['grid_x'] = np.floor(out['x_m'] / grid_size).astype('Int64')
    out['grid_y'] = np.floor(out['y_m'] / grid_size).astype('Int64')
    out['grid_id'] = out['grid_x'].astype(str) + "_" + out['grid_y'].astype(str)

    return out
```

```python
GRID_SIZE = 500  # 필요하면 250, 1000으로 바꿔도 됨

shelter_g = add_grid_key(shelter, grid_size=GRID_SIZE)
store_g = add_grid_key(store, grid_size=GRID_SIZE)
hospital_g = add_grid_key(hospital, grid_size=GRID_SIZE)
station_master_g = add_grid_key(station_master_merge, grid_size=GRID_SIZE)
subway_time_g = add_grid_key(subway_time_merged, grid_size=GRID_SIZE)
```

```python
def check_grid_mapping(df, name):
    missing = df['grid_id'].isna().sum()
    print(f"{name}: total={len(df)}, grid_id missing={missing}")

check_grid_mapping(shelter_g, "shelter")
check_grid_mapping(store_g, "store")
check_grid_mapping(hospital_g, "hospital")
check_grid_mapping(station_master_g, "station_master")
check_grid_mapping(subway_time_g, "subway_time")
```

```python
#지하철 역 부분만 다시 수정하기
import re
import numpy as np
import pandas as pd

def clean_station_name(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    x = re.sub(r'\s+', '', x)          # 공백 제거
    x = re.sub(r'\(.*?\)', '', x)      # 괄호 내용 제거
    return x

def clean_line_name(x):
    if pd.isna(x):
        return np.nan
    x = str(x).strip()
    x = re.sub(r'\s+', '', x)

    # 자주 나오는 호선명 표준화
    line_map = {
        '9호선2~3단계': '9호선',
        '9호선2단계': '9호선',
        '9호선3단계': '9호선',
    }
    return line_map.get(x, x)

# 복사
station_master_merge2 = station_master_raw[['역사명', '호선', '위도', '경도']].copy()
station_master_merge2 = station_master_merge2.rename(columns={
    '역사명': 'station_name',
    '호선': 'line_name',
    '위도': 'lat',
    '경도': 'lon'
})

subway_time2 = subway_time_raw.copy().rename(columns={
    '지하철역': 'station_name',
    '호선명': 'line_name'
})

# 정제 컬럼 생성
station_master_merge2['station_name_clean'] = station_master_merge2['station_name'].apply(clean_station_name)
station_master_merge2['line_name_clean'] = station_master_merge2['line_name'].apply(clean_line_name)

subway_time2['station_name_clean'] = subway_time2['station_name'].apply(clean_station_name)
subway_time2['line_name_clean'] = subway_time2['line_name'].apply(clean_line_name)

# merge
subway_time_merged2 = subway_time2.merge(
    station_master_merge2[['station_name_clean', 'line_name_clean', 'lat', 'lon']],
    on=['station_name_clean', 'line_name_clean'],
    how='left'
)

# 결과 확인
unmatched_after = subway_time_merged2['lat'].isna().sum()
print("정제 후 위경도 매핑 안 된 행 수:", unmatched_after)
print("전체 행 수:", len(subway_time_merged2))
print("매핑 성공률:", round((1 - unmatched_after / len(subway_time_merged2)) * 100, 2), "%")

display(
    subway_time_merged2[subway_time_merged2['lat'].isna()][['line_name', 'station_name']]
    .drop_duplicates()
    .sort_values(['line_name', 'station_name'])
    .head(30)
)
```

```python
# 시간대별로 압축
time_cols = subway_time_merged2.columns

def sum_time_range(df, start, end, prefix='승차인원'):
    cols = [c for c in df.columns if prefix in c and any(f"{h:02d}시" in c for h in range(start, end))]
    return df[cols].sum(axis=1)

subway_time_feat = subway_time_merged2.copy()

# 출근
subway_time_feat['commute_in'] = sum_time_range(subway_time_feat, 7, 10, '승차인원')
subway_time_feat['commute_out'] = sum_time_range(subway_time_feat, 7, 10, '하차인원')

# 낮
subway_time_feat['day_in'] = sum_time_range(subway_time_feat, 10, 17, '승차인원')
subway_time_feat['day_out'] = sum_time_range(subway_time_feat, 10, 17, '하차인원')

# 퇴근
subway_time_feat['evening_in'] = sum_time_range(subway_time_feat, 17, 20, '승차인원')
subway_time_feat['evening_out'] = sum_time_range(subway_time_feat, 17, 20, '하차인원')
```

```python
# 겹치는 월 제거
subway_station_level = (
    subway_time_feat
    .groupby(['line_name', 'station_name', 'lat', 'lon'])
    [['commute_in','commute_out','day_in','day_out','evening_in','evening_out']]
    .mean()
    .reset_index()
)
```

```python
subway_station_level = add_grid_key(subway_station_level, grid_size=500)
```

```python
subway_grid = (
    subway_station_level
    .groupby('grid_id')
    [['commute_in','commute_out','day_in','day_out','evening_in','evening_out']]
    .sum()
    .reset_index()
)
```

```python
import os
import re
import numpy as np
import pandas as pd

from pyproj import Transformer
from scipy.spatial import cKDTree

base_path = "/content/drive/MyDrive/승화 토프"
processed_path = os.path.join(base_path, "processed")

# 250m 생활인구
life_pop_path = os.path.join(base_path, "250m_생활인구.csv")
life_pop_df = pd.read_csv(life_pop_path, encoding="cp949")

# processed 파일들
shelter_df = pd.read_csv(os.path.join(processed_path, "shelter_processed.csv"), encoding="utf-8-sig")
store_df = pd.read_csv(os.path.join(processed_path, "store_processed.csv"), encoding="utf-8-sig")
hospital_df = pd.read_csv(os.path.join(processed_path, "hospital_processed.csv"), encoding="utf-8-sig")
subway_df = pd.read_csv(os.path.join(processed_path, "subway_time_processed.csv"), encoding="utf-8-sig")

print("life_pop_df:", life_pop_df.shape)
print("shelter_df:", shelter_df.shape)
print("store_df:", store_df.shape)
print("hospital_df:", hospital_df.shape)
print("subway_df:", subway_df.shape)
```

```python
# 250m 생활인구의 격자코드만 추출
life_grid = (
    life_pop_df[['250M격자']]
    .drop_duplicates()
    .rename(columns={'250M격자': '격자코드'})
    .reset_index(drop=True)
)

print("고유 250m 격자 수:", len(life_grid))
display(life_grid.head())
```

```python
# 변환기
to_lonlat = Transformer.from_crs("EPSG:5179", "EPSG:4326", always_xy=True)

def gridcode_to_xy_dasa(code: str):
    """
    서울 '다사' 격자코드를 250m 격자 중심좌표(EPSG:5179)로 변환
    예: 다사35005075 -> x=935125, y=1950875 형태
    """
    code = str(code).strip()
    if not code.startswith("다사"):
        return np.nan, np.nan

    digits = code[2:]
    if len(digits) != 8 or not digits.isdigit():
        return np.nan, np.nan

    x_part = int(digits[:4])   # 예: 3500
    y_part = int(digits[4:])   # 예: 5075

    # '다사' 100km 블록 기준 250m 격자 중심점
    # 서울 예시 코드들과 맞는 기준
    x = 900125 + x_part * 10
    y = 1900125 + y_part * 10
    return x, y
```

```python
# 마스터 250m 격자 좌표 생성
life_grid[['x_m', 'y_m']] = life_grid['격자코드'].apply(
    lambda x: pd.Series(gridcode_to_xy_dasa(x))
)

# lon, lat 추가
lon, lat = to_lonlat.transform(life_grid['x_m'].values, life_grid['y_m'].values)
life_grid['lon'] = lon
life_grid['lat'] = lat

print(life_grid.shape)
display(life_grid.head())
```

```python
# 좌표 이상 여부 확인
print(life_grid[['lon', 'lat']].describe())
print("좌표 결측 수:")
print(life_grid[['x_m', 'y_m', 'lon', 'lat']].isnull().sum())
```

```python
# 점 데이터 가까운 250m 격자에 매핑하기
# 250m 마스터 KDTree
grid_tree = cKDTree(life_grid[['x_m', 'y_m']].values)

def attach_to_250m_grid(point_df, lon_col='lon', lat_col='lat', max_dist_m=180):
    """
    point_df의 점 데이터를 가장 가까운 250m 격자코드에 붙임
    max_dist_m보다 멀면 매핑 실패 처리
    """
    out = point_df.copy()

    # lon, lat -> x_m, y_m
    if 'x_m' not in out.columns or 'y_m' not in out.columns:
        tr = Transformer.from_crs("EPSG:4326", "EPSG:5179", always_xy=True)
        x, y = tr.transform(out[lon_col].values, out[lat_col].values)
        out['x_m'] = x
        out['y_m'] = y

    dists, idxs = grid_tree.query(out[['x_m', 'y_m']].values, k=1)

    out['nearest_dist_m'] = dists
    out['격자코드'] = life_grid.iloc[idxs]['격자코드'].values

    out.loc[out['nearest_dist_m'] > max_dist_m, '격자코드'] = np.nan
    return out
```

```python
shelter_map = attach_to_250m_grid(shelter_df, lon_col='lon', lat_col='lat', max_dist_m=180)
store_map = attach_to_250m_grid(store_df, lon_col='lon', lat_col='lat', max_dist_m=180)
hospital_map = attach_to_250m_grid(hospital_df, lon_col='lon', lat_col='lat', max_dist_m=180)
subway_map = attach_to_250m_grid(subway_df, lon_col='lon', lat_col='lat', max_dist_m=180)

print("대피소 매핑 성공률:", round(shelter_map['격자코드'].notna().mean() * 100, 2), "%")
print("상권 매핑 성공률:", round(store_map['격자코드'].notna().mean() * 100, 2), "%")
print("병원 매핑 성공률:", round(hospital_map['격자코드'].notna().mean() * 100, 2), "%")
print("지하철 매핑 성공률:", round(subway_map['격자코드'].notna().mean() * 100, 2), "%")
```

```python
# 생활인구 전처리
life_pop = life_pop_df.copy()
life_pop = life_pop.rename(columns={'250M격자': '격자코드'})

# 숫자형 변환
num_cols = [c for c in life_pop.columns if c not in ['일자', '시간', '행정동코드', '격자코드']]

for c in num_cols:
    life_pop[c] = pd.to_numeric(life_pop[c].replace('*', np.nan), errors='coerce')

life_pop['시간'] = pd.to_numeric(life_pop['시간'], errors='coerce')
life_pop['일자'] = pd.to_datetime(life_pop['일자'].astype(str), format='%Y%m%d', errors='coerce')

print(life_pop.shape)
display(life_pop.head())
```

```python
# 1번 데이터셋 : 인구 + 시설 데이터셋
life_pop_feat = life_pop.copy()

life_pop_feat['출근시간대'] = life_pop_feat['시간'].between(7, 9)
life_pop_feat['주간시간대'] = life_pop_feat['시간'].between(10, 17)
life_pop_feat['퇴근시간대'] = life_pop_feat['시간'].between(18, 20)
life_pop_feat['야간시간대'] = ~(
    life_pop_feat['출근시간대'] |
    life_pop_feat['주간시간대'] |
    life_pop_feat['퇴근시간대']
)

pop_base = life_pop_feat.groupby('격자코드', as_index=False).agg(
    생활인구_평균=('생활인구합계', 'mean'),
    생활인구_최대=('생활인구합계', 'max'),
    생활인구_최소=('생활인구합계', 'min')
)

commute_pop = (
    life_pop_feat[life_pop_feat['출근시간대']]
    .groupby('격자코드', as_index=False)['생활인구합계']
    .mean()
    .rename(columns={'생활인구합계': '생활인구_출근시간평균'})
)

day_pop = (
    life_pop_feat[life_pop_feat['주간시간대']]
    .groupby('격자코드', as_index=False)['생활인구합계']
    .mean()
    .rename(columns={'생활인구합계': '생활인구_주간평균'})
)

evening_pop = (
    life_pop_feat[life_pop_feat['퇴근시간대']]
    .groupby('격자코드', as_index=False)['생활인구합계']
    .mean()
    .rename(columns={'생활인구합계': '생활인구_퇴근시간평균'})
)

night_pop = (
    life_pop_feat[life_pop_feat['야간시간대']]
    .groupby('격자코드', as_index=False)['생활인구합계']
    .mean()
    .rename(columns={'생활인구합계': '생활인구_야간평균'})
)

demo_cols = [c for c in life_pop_feat.columns if c.startswith('남자') or c.startswith('여자')]
demo_feat = life_pop_feat.groupby('격자코드', as_index=False)[demo_cols].mean()

population_feat = pop_base.merge(commute_pop, on='격자코드', how='left')
population_feat = population_feat.merge(day_pop, on='격자코드', how='left')
population_feat = population_feat.merge(evening_pop, on='격자코드', how='left')
population_feat = population_feat.merge(night_pop, on='격자코드', how='left')
population_feat = population_feat.merge(demo_feat, on='격자코드', how='left')

print(population_feat.shape)
display(population_feat.head())
```

```python
# 대피소
shelter_feat = (
    shelter_map.dropna(subset=['격자코드'])
    .groupby('격자코드', as_index=False)
    .agg(
        대피소수=('시설명', 'count'),
        대피소면적합=('시설면적', 'sum'),
        대피소유형수=('구분명', 'nunique')
    )
)

# 병원
hospital_map['응급실운영여부(1/2)'] = pd.to_numeric(hospital_map['응급실운영여부(1/2)'], errors='coerce')
hospital_map['응급병원여부'] = (hospital_map['응급실운영여부(1/2)'] == 1).astype(int)

hospital_feat = (
    hospital_map.dropna(subset=['격자코드'])
    .groupby('격자코드', as_index=False)
    .agg(
        병원수=('기관ID', 'count'),
        응급병원수=('응급병원여부', 'sum'),
        병원종류수=('병원분류명', 'nunique')
    )
)

# 상권
store_feat = (
    store_map.dropna(subset=['격자코드'])
    .groupby('격자코드', as_index=False)
    .agg(
        상권수=('상호명', 'count'),
        상권대분류수=('상권업종대분류명', 'nunique'),
        상권중분류수=('상권업종중분류명', 'nunique')
    )
)
```

```python
population_facility_dataset = life_grid[['격자코드', 'lon', 'lat']].copy()

population_facility_dataset = population_facility_dataset.merge(population_feat, on='격자코드', how='left')
population_facility_dataset = population_facility_dataset.merge(shelter_feat, on='격자코드', how='left')
population_facility_dataset = population_facility_dataset.merge(hospital_feat, on='격자코드', how='left')
population_facility_dataset = population_facility_dataset.merge(store_feat, on='격자코드', how='left')

for c in ['대피소수', '대피소면적합', '대피소유형수',
          '병원수', '응급병원수', '병원종류수',
          '상권수', '상권대분류수', '상권중분류수']:
    if c in population_facility_dataset.columns:
        population_facility_dataset[c] = population_facility_dataset[c].fillna(0)

print("population_facility_dataset:", population_facility_dataset.shape)
display(population_facility_dataset.head())
```

2번 데이터셋 : 지하철 + 노선 + 시간대별 승하차 데이터셋

```python
time_cols = [c for c in subway_map.columns if ('승차인원' in c or '하차인원' in c)]
print("시간대 컬럼 수:", len(time_cols))
print(time_cols[:10])
```

```python
subway_station_level = (
    subway_map.dropna(subset=['격자코드'])
    .groupby(['격자코드', 'line_name', 'station_name'], as_index=False)[time_cols]
    .mean()
)

print(subway_station_level.shape)
display(subway_station_level.head())
```

```python
def get_cols_by_hour_range(cols, start_hour, end_hour, keyword):
    selected = []
    for c in cols:
        if keyword not in c:
            continue
        m = re.match(r'(\d{2})시-\d{2}시', c)
        if m:
            h = int(m.group(1))
            if start_hour <= h <= end_hour:
                selected.append(c)
    return selected

ride_cols = [c for c in time_cols if '승차인원' in c]
alight_cols = [c for c in time_cols if '하차인원' in c]

commute_ride_cols = get_cols_by_hour_range(ride_cols, 7, 9, '승차인원')
commute_alight_cols = get_cols_by_hour_range(alight_cols, 7, 9, '하차인원')

day_ride_cols = get_cols_by_hour_range(ride_cols, 10, 17, '승차인원')
day_alight_cols = get_cols_by_hour_range(alight_cols, 10, 17, '하차인원')

evening_ride_cols = get_cols_by_hour_range(ride_cols, 18, 20, '승차인원')
evening_alight_cols = get_cols_by_hour_range(alight_cols, 18, 20, '하차인원')

subway_station_level['출근_승차합'] = subway_station_level[commute_ride_cols].sum(axis=1)
subway_station_level['출근_하차합'] = subway_station_level[commute_alight_cols].sum(axis=1)
subway_station_level['주간_승차합'] = subway_station_level[day_ride_cols].sum(axis=1)
subway_station_level['주간_하차합'] = subway_station_level[day_alight_cols].sum(axis=1)
subway_station_level['퇴근_승차합'] = subway_station_level[evening_ride_cols].sum(axis=1)
subway_station_level['퇴근_하차합'] = subway_station_level[evening_alight_cols].sum(axis=1)
```

```python
def join_unique_sorted(series):
    vals = sorted(set([str(x) for x in series.dropna().astype(str)]))
    return ", ".join(vals)

agg_dict = {c: 'sum' for c in time_cols}
agg_dict.update({
    '출근_승차합': 'sum',
    '출근_하차합': 'sum',
    '주간_승차합': 'sum',
    '주간_하차합': 'sum',
    '퇴근_승차합': 'sum',
    '퇴근_하차합': 'sum',
    'line_name': join_unique_sorted,
    'station_name': join_unique_sorted
})

subway_grid_dataset = (
    subway_station_level
    .groupby('격자코드', as_index=False)
    .agg(agg_dict)
)

station_count = (
    subway_station_level.groupby('격자코드', as_index=False)['station_name']
    .nunique()
    .rename(columns={'station_name': '지하철역수'})
)

line_count = (
    subway_station_level.groupby('격자코드', as_index=False)['line_name']
    .nunique()
    .rename(columns={'line_name': '지하철노선수'})
)

subway_grid_dataset = subway_grid_dataset.merge(station_count, on='격자코드', how='left')
subway_grid_dataset = subway_grid_dataset.merge(line_count, on='격자코드', how='left')

subway_grid_dataset = subway_grid_dataset.rename(columns={
    'line_name': '노선명목록',
    'station_name': '역사명목록'
})

subway_grid_dataset = life_grid[['격자코드', 'lon', 'lat']].merge(
    subway_grid_dataset, on='격자코드', how='right'
)

print("subway_grid_dataset:", subway_grid_dataset.shape)
display(subway_grid_dataset.head())
```

```python
def dataset_overview(df, name):
    print(f"\n{'='*80}")
    print(f"[{name}]")
    print(f"{'='*80}")
    print("shape:", df.shape)
    print("\n[columns]")
    print(df.columns.tolist())

    print("\n[dtypes]")
    display(df.dtypes.to_frame("dtype"))

    print("\n[missing values]")
    missing = df.isnull().sum().sort_values(ascending=False)
    missing_ratio = (df.isnull().mean() * 100).sort_values(ascending=False)
    missing_df = pd.DataFrame({
        "missing_count": missing,
        "missing_ratio(%)": missing_ratio.round(2)
    })
    display(missing_df[missing_df["missing_count"] > 0])

    print("\n[numeric summary]")
    display(df.describe())

dataset_overview(population_facility_dataset, "population_facility_dataset")
dataset_overview(subway_grid_dataset, "subway_grid_dataset")
```
